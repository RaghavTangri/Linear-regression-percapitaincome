```python
# load the librires
import pandas as pd
```

```python
# create the dataframe
df=pd.read_csv('canada.csv')
df.head()
```

```python
x=df.iloc[:,0:1].values
y=df.iloc[:,1].values
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='red')
```

```python
# split the data
# 70% for trainign and 30% for testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=5)
```

```python
# train using simple linear regression model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# make predictrions based on test data
y_pred=model.predict(x_test)
y_pred
```

```python
# find the r squared value by comparing test data and predicted data
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
# another way to know the score of a linear regression model
model.score(x_test,y_test)
```

```python
# preduct the per capita income during the years 2020 and 2021
model.predict([[2020],[2021]])
```

```python
# draw scatter plot and regresion line
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,model.predict(x_train),color='blue')
plt.title('PER CAIPTA INCOME OF CANADA')
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.show()
```
